```python
import json
```

```python
with open('data/project-1-at-2022-10-05-12-17-625dbd51.json') as f:
    data = json.load(f)
```

```python
def walk(node):
    if isinstance(node, list):
        for item in node:
            walk(item)
    elif isinstance(node, dict):
        for key, item in node.items():
            if key == 'frame':
                node[key] = int(item * 30/100)
            
            walk(item)
```
